<h1>Modification du fichier ticket

```python
import datetime
```

```python
df_ticket['search_ts']= pd.to_datetime(df_ticket['search_ts']) 
df_ticket['departure_ts']= pd.to_datetime(df_ticket['departure_ts']) 
df_ticket['arrival_ts']= pd.to_datetime(df_ticket['arrival_ts']) 
```

```python
def jour_semaine(date):
    return date.day_name()
```

```python
df_ticket['Weekend_search']=df_ticket['search_ts'].apply(jour_semaine)
df_ticket['dep_ts_weekend']=df_ticket['departure_ts'].apply(jour_semaine)
```

```python
df_ticket
```

```python
df_ticket['search_ts']
```

```python
df_ticket['day_night_search']=df_ticket['search_ts'].dt.hour
df_ticket['day_night_dep']=df_ticket['departure_ts'].dt.hour
```

```python
df_ticket
```

```python
def calcul_time(dat_ar,dat_d):
    dure_min=dat_ar-dat_d
    return((dure_min.dt.total_seconds()//60)%60)
```

```python
df_ticket['Duree_min']=calcul_time(df_ticket['arrival_ts'],df_ticket['departure_ts'])
```

```python
#lat_o=float(df_city[df_city['id']==dat_ticket['o_city']]['latitude'].values)
def found_lat_city(x):
    return float(df_city[df_city['id']==x]['latitude'])
def found_long_city(x):
    return float(df_city[df_city['id']==x]['longitude'])
df_ticket['lat_o']=df_ticket['o_city'].apply(found_lat_city)
df_ticket['long_o']=df_ticket['o_city'].apply(found_long_city)
```

```python
df_ticket['lat_d']=df_ticket['d_city'].apply(found_lat_city)
df_ticket['long_d']=df_ticket['d_city'].apply(found_long_city)
```

```python
df_ticket['lat_d']=pd.to_numeric(df_ticket['lat_d'])
df_ticket['long_d']=pd.to_numeric(df_ticket['long_d'])
df_ticket['lat_o']=pd.to_numeric(df_ticket['lat_o'])
df_ticket['long_o']=pd.to_numeric(df_ticket['long_o'])
```

```python
from geopy.distance import geodesic
def calcul_distance(row):
    coords_1 = (row['lat_d'],row['long_d'])
    coords_2 = (row['lat_o'], row['long_o'])
    return geodesic(coords_1, coords_2).km
df_ticket['Distance_km']=df_ticket.apply(calcul_distance,axis=1)
```

```python
def recherche_transport(x): 
    return (df_providers[df_providers['id']==x]['transport_type'].values[0])
df_ticket['type_transp']=df_ticket['company'].apply(recherche_transport)
```

```python
from geopy.geocoders import Nominatim
```

```python
geolocator = Nominatim(user_agent="geoapiExercises")
country=[]
city=[]
for Latitude,Longitude in zip(df_city['latitude'],df_city['longitude']):
    Latitude=str(Latitude)
    Longitude =str(Longitude)
    location = geolocator.reverse(Latitude+","+Longitude)
    try:
        address = location.raw['address']
        pays=address.get('country')
        cite=address.get('city')
    except AttributeError:
        pays='NO COUNTRY'
        cite='NO CITY'
    country.append(pays)
    city.append(cite)
```

```python
df_city['country']=country
df_city['city']=city
```
